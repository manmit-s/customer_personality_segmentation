## **2. Exploring Data**

```python
df.Dt_Customer = pd.to_datetime(df.Dt_Customer, format = '%d-%m-%Y')
df.Dt_Customer
```

```python
#define numerical and categorical columns
#drop the ID column
columns = df.drop(columns=['ID']).columns.to_list()
columns
```

```python
numeric_features = df.select_dtypes(include = 'number').columns.to_list()
categorical_features = df.select_dtypes(include = 'object').columns.tolist()

print(f"We have {len(numeric_features)} numeric features - {numeric_features}")
print(f"We have {len(categorical_features)} numeric features - {categorical_features}")
```

```python
df[categorical_features[0]].nunique()
```

```python
df[categorical_features[0]].value_counts()
```

```python
# prportion of count data on categorical columns
for col in categorical_features:
    print(df[col].value_counts(normalize=True)*100)
    print("------------------------------")
```

## **Univariate Analysis**

### Numerical Features

```python
plt.figure(figsize = (15, 10))
plt.suptitle("Univariate Analysis of Numerical Features", fontsize = 20, fontweight = 'bold', alpha = 0.8, y = 1.)

for i in range(0, len(numeric_features)):
    plt.subplot(7, 4, i+1)
    sns.kdeplot(x = df[numeric_features[i]], color = 'blue')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()

#save plot
plt.savefig("images/Univariate_Num.png")
```

```python
df.Z_CostContact.plot()
```

```python
df.Z_Revenue.plot()
```

```python
# as these columns dont have any variance we can drop these
df = df.drop(columns = ['Z_CostContact', 'Z_Revenue'])
df.columns
```

### Categorial Features

```python
#categorical columns
plt.figure(figsize = (15, 8))
plt.suptitle("Univariate Analysis of Categorical Features", fontsize = 20, fontweight = 'bold', alpha = 0.8, y = 1.)

for i in range(0, len(categorical_features)):
    plt.subplot(2, 2, i+1)
    ax = sns.countplot (x=df[categorical_features[i]])
    if df[categorical_features[i]].value_counts().count() <10:  
        for p in ax.patches:
            ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.xlabel(categorical_features[i])
plt.tight_layout()

#save plot
plt.savefig("images/Univariate_Cat.png")
```

```python
df1 = df.copy()
df1['Year'] = df1.Dt_Customer.dt.year
customer_joined_per_year = df1.groupby(by="Year", axis=0)[['ID']].count().reset_index()
customer_joined_per_year.columns = ['Year', 'CUSTOMER_JOINED']
sns.barplot(data=customer_joined_per_year, x='Year', y='CUSTOMER_JOINED')
```
